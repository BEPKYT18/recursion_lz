```python
import random

def labirinth():
    m=random.randint(1,10)
    n=random.randint(1,10)
    lab=[]
    for i in range(0,m):
        lab.append(list())
        for j in range(0,n):
            lab[i].append(random.randint(0,1))
        print(lab[i],'\n')
    return lab


def find_path(lab, start, stop):
    exit=[]
    if start == stop:
        return exit.append(start)
    
    elif 0==[lab[start[0]+1][start[1]]]:
        t=(start[0]+1,start[1])
        return exit.append(t),find_path(lab, t, stop)
    
    if 0==[lab[start[0]][start[1]+1]]:
        t=(start[0],start[1]+1)
        return exit.append(t),find_path(lab, t, stop)
    
    if 0==[lab[start[0]-1][start[1]]]:
        t=(start[0]-1,start[1])
        return exit.append(t),find_path(lab, t, stop)
    
    else:
        return None


def main():
    maze=labirinth()
    x=int(input())
    y=int(input())
    a = (x,y)
    x=int(input())
    y=int(input())
    b = (x,y)
    exit_path=find_path(maze,a,b)
    print(exit_path)

if __name__=='__main__':
    main()
```
